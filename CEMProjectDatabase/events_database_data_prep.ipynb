```python
import openpyxl
import csv
import pandas as pd

# Open the Excel workbook and select the active sheet
excel = openpyxl.load_workbook("Edited_MasterDataList.xlsx")
sheet = excel.active

# Write the data to a CSV file
with open("data.csv", 'w', newline="") as csvfile:
    col = csv.writer(csvfile)
    for r in sheet.rows:
        col.writerow([cell.value for cell in r])

# Read the CSV file into a pandas DataFrame and drop columns with Unnamed header or empty values
df = pd.read_csv("data.csv").dropna(axis=1, how='all')

# if the values are NaN, turn them into 
df = df.fillna(0)
df = df[(df != 0).any(axis=1)]

# Write the cleaned DataFrame back to a CSV file
df.to_csv("cleaned_data.csv", index=False)

print("Cleaned data has been written to cleaned_data.csv")
```

```python
column_names = ["EventName", "EventType", "EventYear", "OrgName", "OnSite", "OffSite"]
data = {"EventName": None,
        "EventType": None,
        "EventYear": None,
        "OrgName": None,
        "OnSite": None,
        "OffSite": None
       }
new_df = pd.DataFrame(data, columns=column_names)

original_columns = df.columns

for index, row in df.iloc[1:].iterrows():
    for i in range(1, len(original_columns), 2):
        broken = original_columns[i].split()
        event_type = broken[1]
        event_year = broken[0]
        new_row = {
            "EventName": original_columns[i],
            "EventType": event_type,
            "EventYear": event_year,
            "OrgName": row["PK-12 COLLABORATIVE Registration / Attendance Data 2018-Present"],
            "OnSite": row[original_columns[i]],
            "OffSite": row[original_columns[i] + ".1"]
        }
        new_df.loc[len(new_df.index)] = new_row 
        
new_df = new_df.drop(0, axis=0)

            
```
